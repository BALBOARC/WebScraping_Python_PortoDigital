```python
soup = BeautifulSoup(page, 'html5lib')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
empresa = soup.find_all('p', class_='MsoNormal')
print(empresa)
```

```python
for item in empresa:
    nome_empresa = item.text.split()
    link_empresa = []
    link_empresa.extend(item.find_all('a'))
    print(nome_empresa)
    print(link_empresa)
```
